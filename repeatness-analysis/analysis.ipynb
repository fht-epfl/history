```python
# repeatness_analysis.py

import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, leaves_list

def plot_hierarchical_heatmap(df_ima, 
                              unit: str = 'book',
                              big: str  = 'big_label',
                              small: str = 'small_label',
                              method: str = 'grouped',
                              cmap: str = 'viridis',
                              figsize=(16,8)):
    """
    method:
      - 'grouped': 先依 big_label 排序，再依 small_label 排序，畫出分組 heatmap
      - 'cluster': 對 small_label row 做階層聚類，依照葉子順序重排列
    """
    # 1. Pivot table with MultiIndex rows (big, small)
    pivot = (df_ima
             .groupby([big, small, unit])
             .size()
             .unstack(unit, fill_value=0))
    
    # 2. Decide row order
    if method == 'grouped':
        # 直接依 big_label, small_label 排序
        ordered = pivot.sort_index(level=[0,1])
        row_labels = ordered.index
    elif method == 'cluster':
        # 階層聚類：先算 linkage，再用 leaves_list 拿出 row order
        Z = linkage(pivot.values, method='average', metric='euclidean')
        order = leaves_list(Z)
        ordered = pivot.iloc[order]
        row_labels = ordered.index
    else:
        raise ValueError("method must be 'grouped' or 'cluster'")

    # 3. Plot heatmap
    mat = ordered.values
    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(mat, aspect='auto', cmap=cmap)
    cbar = fig.colorbar(im, ax=ax)
    cbar.set_label('count')

    # 4. y-axis: 只顯示 small_label，並在不同 big_label 群間畫分隔線
    small_ticks = [lbl[1] for lbl in row_labels]
    ax.set_yticks(np.arange(len(small_ticks)))
    ax.set_yticklabels(small_ticks, fontsize=8)

    # 找出分組邊界
    big_list = [lbl[0] for lbl in row_labels]
    boundaries = []
    for i in range(1, len(big_list)):
        if big_list[i] != big_list[i-1]:
            boundaries.append(i-0.5)
    for b in boundaries:
        ax.axhline(b, color='white', linewidth=2)

    # 5. 在右側新增一條 axis，標出每個 big_label 的中間位置
    unique_bigs = []
    positions = []
    start = 0
    current = big_list[0]
    for i, b in enumerate(big_list + [None]):
        if b != current:
            unique_bigs.append(current)
            end = i
            positions.append((start + end-1)/2)
            current = b
            start = i
    ax2 = ax.twinx()
    ax2.set_yticks(positions)
    ax2.set_yticklabels(unique_bigs, fontsize=9)
    ax2.set_ylim(ax.get_ylim())
    ax2.tick_params(length=0)

    # 6. x-axis: 書名
    ax.set_xticks(np.arange(mat.shape[1]))
    ax.set_xticklabels(ordered.columns, rotation=90, fontsize=8)

    ax.set_title(f"Hierarchical heatmap ({method}) of {small} within {big} × {unit}")
    plt.tight_layout()
    plt.show()

plot_hierarchical_heatmap(df_ima, method='grouped')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, leaves_list
from matplotlib.colors import Normalize, LogNorm, PowerNorm, SymLogNorm

def plot_hierarchical_heatmap_text(df_ima: pd.DataFrame,
                                   df_books: pd.DataFrame,
                                   big: str = 'big_label',
                                   small: str = 'small_label',
                                   unit: str = 'book',
                                   method: str = 'grouped',
                                   cmap: str = 'viridis',
                                   scale: str = 'linear',
                                   linthresh: float = 1.0,
                                   gamma: float = 0.5,
                                   figsize=(16,8)):
    """
    基于文本统计每个 small_label 的“词汇池”在各本书中出现的总次数，
    并绘制 hierarchical heatmap。

    参数
    ----
    df_ima: 包含 columns [word, big_label, small_label, book]
    df_books: 包含 columns [book, year, text_chunk_smallest]
    method: 'grouped' 或 'cluster'
    cmap: matplotlib colormap 名称
    scale: 颜色映射尺度, 支持 'linear', 'log', 'sqrt', 'symlog'
    linthresh: 对于 'symlog'，线性阈值
    gamma: 对于 'sqrt' (PowerNorm)，gamma 值
    """
    # 1. small_label -> 词表
    label2words = df_ima.groupby(small)['word'].apply(list).to_dict()

    # 2. 按年份排序书名
    book_order = (df_books[['book','year']]
                  .drop_duplicates('book')
                  .sort_values('year')['book']
                  .tolist())

    # 3. 逐书、逐标签计数
    recs = []
    for book in book_order:
        sentences = df_books.loc[df_books.book==book, 'text_chunk_smallest'].iloc[0]
        full_text = ''.join(sentences)
        for bl, sl in df_ima[[big, small]].drop_duplicates().values:
            count = sum(full_text.count(w) for w in label2words.get(sl, []))
            recs.append({unit: book, big: bl, small: sl, 'count': count})
    df_cnt = pd.DataFrame(recs)

    # 4. pivot
    pivot = (df_cnt
             .pivot_table(index=[big, small],
                          columns=unit,
                          values='count',
                          fill_value=0))

    # 5. 排序行
    if method == 'grouped':
        ordered = pivot.sort_index(level=[0,1])
    elif method == 'cluster':
        Z = linkage(pivot.values, method='average', metric='euclidean')
        order = leaves_list(Z)
        ordered = pivot.iloc[order]
    else:
        raise ValueError("method must be 'grouped' or 'cluster'")

    row_labels = ordered.index
    mat = ordered.values

    # 6. 选择颜色归一化方式
    if scale == 'linear':
        norm = Normalize(vmin=mat.min(), vmax=mat.max())
    elif scale == 'log':
        # 避免 0 值, 添加微小常数
        norm = LogNorm(vmin=max(mat.min(), 1e-3), vmax=mat.max())
    elif scale == 'sqrt':
        norm = PowerNorm(gamma=gamma, vmin=mat.min(), vmax=mat.max())
    elif scale == 'symlog':
        norm = SymLogNorm(linthresh=linthresh, vmin=mat.min(), vmax=mat.max())
    else:
        raise ValueError("scale must be 'linear','log','sqrt', or 'symlog'")

    # 7. 绘图
    fig, ax = plt.subplots(figsize=figsize)
    im = ax.imshow(mat, aspect='auto', cmap=cmap, norm=norm)
    fig.colorbar(im, ax=ax, label='count')

    # y ticks
    small_ticks = [sl for _, sl in row_labels]
    ax.set_yticks(np.arange(len(small_ticks)))
    ax.set_yticklabels(small_ticks, fontsize=8)

    # 分割线 & right big labels
    big_list = [bl for bl, _ in row_labels]
    boundaries = [i-0.5 for i in range(1,len(big_list)) if big_list[i]!=big_list[i-1]]
    for b in boundaries:
        ax.axhline(b, color='white', linewidth=2)
    unique_bigs, positions = [], []
    start, cur = 0, big_list[0]
    for i, b in enumerate(big_list + [None]):
        if b != cur:
            unique_bigs.append(cur)
            positions.append((start + i - 1)/2)
            start, cur = i, b
    ax2 = ax.twinx()
    ax2.set_yticks(positions)
    ax2.set_yticklabels(unique_bigs, fontsize=9)
    ax2.set_ylim(ax.get_ylim())
    ax2.tick_params(length=0)

        # x ticks: show book name and year
    years = df_books.set_index('book')['year'].to_dict()
    labels = [f"{b}\n{years.get(b,'')}" for b in book_order]
    ax.set_xticks(np.arange(len(book_order)))
    ax.set_xticklabels(labels, rotation=90, fontsize=8)

    ax.set_title(f"Hierarchical heatmap ({method}, scale={scale}) of '{small}' in books")
    plt.tight_layout()
    plt.show()

# 1. 先筛出朱天心的书
zh_books = df_books[(df_books.author == '朱天心') & (df_books.year >= 1987)]

# 2. 再从 df_ima 里，只保留这些书对应的意象
zh_ima = df_ima[df_ima.book.isin(zh_books.book)]

# 3. 然后调用热图函数
plot_hierarchical_heatmap_text(
    df_ima=zh_ima,
    df_books=zh_books,
    method='grouped',   # 或 'cluster'
    scale='sqrt'        # 例如平方根刻度
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, leaves_list
from matplotlib.colors import Normalize, LogNorm, PowerNorm, SymLogNorm


def plot_hierarchical_heatmap_text(df_ima: pd.DataFrame,
                                   df_books: pd.DataFrame,
                                   big: str = 'big_label',
                                   small: str = 'small_label',
                                   unit: str = 'book',
                                   method: str = 'grouped',
                                   cmap: str = 'viridis',
                                   scale: str = 'linear',
                                   linthresh: float = 1.0,
                                   gamma: float = 0.5,
                                   figsize=(20,12)):
    """
    绘制分组或聚类的 hierarchical heatmap，y 轴分组中留空行，
    左侧显示英文小标签，右侧显示英文大标签，并在底部横放 colorbar。
    """
    # 1. 词汇池
    df_ima = df_ima[df_ima.apply(lambda x: len(x['word']) > 1, axis=1)]
    label2words = df_ima.groupby(small)['word'].apply(list).to_dict()
    # 2. 排序书
    book_order = df_books[['book','year']].drop_duplicates('book').sort_values('year')['book'].tolist()
    # 3. 计数
    recs = []
    for book in book_order:
        text = ''.join(df_books.loc[df_books.book==book,'text_chunk_smallest'].iloc[0])
        for bl, sl in df_ima[[big,small]].drop_duplicates().values:
            cnt = sum(text.count(w) for w in label2words.get(sl,[]))
            recs.append({unit:book,big:bl,small:sl,'count':cnt})
    df_cnt = pd.DataFrame(recs)
    pivot = df_cnt.pivot_table(index=[big,small],columns=unit,values='count',fill_value=0)
    # 4. 排序
    if method=='grouped':
        ordered = pivot.sort_index(level=[0,1])
    else:
        Z=linkage(pivot.values,method='average',metric='euclidean')
        ordered=pivot.iloc[leaves_list(Z)]
    idx = list(ordered.index)
    mat = ordered.values
    # 5. 插入空行
    big_list=[b for b,_ in idx]
    groups = {}
    for i,(b,sl) in enumerate(idx): groups.setdefault(b,[]).append(i)
    new_mat=[]; new_idx=[]
    for b,rows in groups.items():
        for r in rows:
            new_mat.append(mat[r])
            new_idx.append(idx[r])
        # 插入空 row
        new_mat.append([np.nan]*mat.shape[1])
        new_idx.append((b,None))
    mat2 = np.vstack(new_mat)
    # 6. 归一化
    if scale=='linear': norm=Normalize(np.nanmin(mat2),np.nanmax(mat2))
    elif scale=='log': norm=LogNorm(max(np.nanmin(mat2),1e-3),np.nanmax(mat2))
    elif scale=='sqrt': norm=PowerNorm(gamma, np.nanmin(mat2), np.nanmax(mat2))
    else: norm=SymLogNorm(linthresh,np.nanmin(mat2),np.nanmax(mat2))
    # 7. 绘图
    fig,ax=plt.subplots(figsize=figsize)
    im=ax.imshow(mat2,aspect='auto',cmap=cmap,norm=norm)
    # colorbar 横向下方
    cbar=fig.colorbar(im,ax=ax,orientation='horizontal',pad=0.15,aspect=50,fraction=0.05)
    cbar.set_label('count')
    # y ticks
    sl_cn=[sl for _,sl in new_idx]
    sl_en=[english_small_label.get(sl,'') if sl else '' for sl in sl_cn]
    ax.set_yticks(np.arange(len(sl_en)))
    ax.set_yticklabels(sl_en,fontsize=8)
    # big labels on right
    big_cn=[b for b,_ in new_idx]
    uniq=[]; pos=[]; start=0; cur=big_cn[0]
    for i,b in enumerate(big_cn+ [None]):
        if b!=cur:
            uniq.append(cur); pos.append((start+i-1)/2); start=i; cur=b
    big_en=[english_big_label.get(b,'') for b in uniq]
    ax2=ax.twinx(); ax2.set_yticks(pos); ax2.set_yticklabels(big_en,fontsize=9)
    ax2.set_ylim(ax.get_ylim()); ax2.tick_params(length=0)
    # x ticks
    years=df_books.set_index('book')['year'].to_dict()
    labels=[f"{b}\n{years[b]}" for b in book_order]
    ax.set_xticks(np.arange(len(book_order))); ax.set_xticklabels(labels,rotation=90,fontsize=8)
    ax.set_title(f"Hierarchical heatmap ({method}, scale={scale}) of '{small}' in books")
    plt.tight_layout(); plt.show()

    
plot_hierarchical_heatmap_text(
    df_ima=zh_ima,
    df_books=zh_books,
    method='grouped',   # 或 'cluster'
    scale='sqrt'        # 例如平方根刻度
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict, Counter
import re
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

class IntraBookEthnicAnalyzer:
    """
    书内族群意象重复性分析器
    基于文本匹配和窗口统计的方法
    """
    
    def __init__(self, df_ima, df_books):
        """
        初始化分析器
        
        Parameters:
        df_ima (DataFrame): 意象数据
        df_books (DataFrame): 书籍数据
        """
        self.df_ima = df_ima.copy()
        self.df_books = df_books.copy()
        
        # 定义族群相关的标签
        self.ethnic_labels = {
            '本省人': ['语言与族群-本省人'],
            '外省人': ['语言与族群-外省人'], 
            '日本人': ['语言与族群-日本人'],
            '原住民': ['语言与族群-原住民']
        }
        
        print("🔍 族群意象分析器初始化完成")
        print(f"目标族群类别: {list(self.ethnic_labels.keys())}")
    
    def get_book_sentences(self, book_title):
        """
        获取指定书籍的句子列表
        
        Parameters:
        book_title (str): 书名
        
        Returns:
        list: 句子列表
        """
        book_data = self.df_books[self.df_books['book'] == book_title]
        
        if book_data.empty:
            available_books = self.df_books['book'].unique()
            print(f"❌ 找不到书籍: '{book_title}'")
            print(f"可用的书籍: {list(available_books)[:10]}...")  # 只显示前10个
            return []
        
        # 获取text_chunk_smallest列
        sentences = book_data['text_chunk_smallest'].iloc[0]
        
        if isinstance(sentences, list):
            return sentences
        else:
            print(f"⚠️ text_chunk_smallest不是列表格式")
            return []
    
    def get_ethnic_words_for_book(self, book_title):
        """
        获取指定书籍中的族群相关词汇
        同时以 book 和 label 为筛选条件
        
        Parameters:
        book_title (str): 书名
        
        Returns:
        dict: {族群类型: [词汇列表]}
        """
        ethnic_words = {}
        
        for ethnic_type, labels in self.ethnic_labels.items():
            words_for_ethnic = []
            
            for label in labels:
                # 同时以 book 和 label 为筛选条件
                matching_rows = self.df_ima[
                    (self.df_ima['book'] == book_title) & 
                    (self.df_ima['label'] == label)
                ]
                
                if not matching_rows.empty:
                    words_list = matching_rows['word'].unique().tolist()
                    words_for_ethnic.extend(words_list)
            
            ethnic_words[ethnic_type] = list(set(words_for_ethnic))  # 去重
        
        return ethnic_words
    
    def count_words_in_text(self, text, word_list):
        """
        统计词汇在文本中的出现次数
        
        Parameters:
        text (str): 文本内容
        word_list (list): 要搜索的词汇列表
        
        Returns:
        int: 总出现次数
        """
        if not text or not word_list:
            return 0
        
        total_count = 0
        for word in word_list:
            if word and word.strip():  # 确保词汇不为空
                # 使用正则表达式计数，避免子串匹配问题
                count = len(re.findall(re.escape(word), text))
                total_count += count
        
        return total_count
    
    def analyze_ethnic_patterns_in_book(self, book_title, window_size=5):
        """
        分析指定书籍中的族群意象模式
        
        Parameters:
        book_title (str): 书名
        window_size (int): 滑动窗口大小
        
        Returns:
        dict: 分析结果
        """
        print(f"\n📖 分析书籍: {book_title}")
        print(f"🔍 窗口大小: {window_size}句")
        
        # 获取句子列表
        sentences = self.get_book_sentences(book_title)
        if not sentences:
            return {}
        
        print(f"📝 总句子数: {len(sentences)}")
        
        # 获取该书的族群相关词汇
        ethnic_words = self.get_ethnic_words_for_book(book_title)
        
        print(f"👥 各族群词汇数量:")
        total_words = 0
        for ethnic_type, words in ethnic_words.items():
            print(f"   • {ethnic_type}: {len(words)}个词汇")
            if words:  # 只在有词汇时显示前几个例子
                examples = words[:3]
                print(f"     例如: {examples}")
            total_words += len(words)
        
        if total_words == 0:
            print("❌ 该书中未找到族群相关词汇")
            return {}
        
        # 滑动窗口分析
        window_analysis = self._sliding_window_analysis_with_grep(
            sentences, ethnic_words, window_size
        )
        
        # 句子级别的统计
        sentence_level_stats = self._analyze_sentence_level_patterns(
            sentences, ethnic_words
        )
        
        # 计算重复性指标
        repeatness_metrics = self._calculate_repeatness_metrics(
            sentence_level_stats, window_analysis, len(sentences)
        )
        
        # 时间序列分析
        temporal_patterns = self._analyze_temporal_patterns(
            sentence_level_stats, len(sentences)
        )
        
        results = {
            'book_title': book_title,
            'total_sentences': len(sentences),
            'ethnic_words': ethnic_words,
            'sentence_level_stats': sentence_level_stats,
            'window_analysis': window_analysis,
            'repeatness_metrics': repeatness_metrics,
            'temporal_patterns': temporal_patterns
        }
        
        return results
    
    def _sliding_window_analysis_with_grep(self, sentences, ethnic_words, window_size):
        """
        使用滑动窗口和文本匹配进行分析
        """
        window_results = []
        total_sentences = len(sentences)
        
        for start_idx in range(0, total_sentences, window_size):
            end_idx = min(start_idx + window_size, total_sentences)
            
            # 获取当前窗口的文本
            window_sentences = sentences[start_idx:end_idx]
            window_text = ' '.join(window_sentences)
            
            # 统计各族群在当前窗口的出现次数
            ethnic_counts = {}
            total_ethnic_mentions = 0
            
            for ethnic_type, words in ethnic_words.items():
                count = self.count_words_in_text(window_text, words)
                ethnic_counts[ethnic_type] = count
                total_ethnic_mentions += count
            
            window_info = {
                'window_start': start_idx,
                'window_end': end_idx - 1,
                'window_center': (start_idx + end_idx - 1) / 2,
                'window_sentences': window_sentences,
                'total_ethnic_mentions': total_ethnic_mentions,
                'ethnic_counts': ethnic_counts,
                'ethnic_diversity': len([c for c in ethnic_counts.values() if c > 0]),
                'dominant_ethnic': max(ethnic_counts.items(), key=lambda x: x[1])[0] if total_ethnic_mentions > 0 else None
            }
            
            # 为每个族群类型添加具体计数
            for ethnic_type in self.ethnic_labels.keys():
                window_info[f'{ethnic_type}_count'] = ethnic_counts.get(ethnic_type, 0)
            
            window_results.append(window_info)
        
        return window_results
    
    def _analyze_sentence_level_patterns(self, sentences, ethnic_words):
        """
        分析句子级别的族群意象分布
        """
        sentence_stats = []
        
        for idx, sentence in enumerate(sentences):
            sentence_ethnic_counts = {}
            total_count = 0
            
            for ethnic_type, words in ethnic_words.items():
                count = self.count_words_in_text(sentence, words)
                sentence_ethnic_counts[ethnic_type] = count
                total_count += count
            
            if total_count > 0:  # 只记录有族群意象的句子
                sentence_stats.append({
                    'sentence_idx': idx,
                    'sentence': sentence,
                    'total_ethnic_count': total_count,
                    'ethnic_counts': sentence_ethnic_counts,
                    'dominant_ethnic': max(sentence_ethnic_counts.items(), key=lambda x: x[1])[0] if total_count > 0 else None
                })
        
        return sentence_stats
    
    def _calculate_repeatness_metrics(self, sentence_stats, window_analysis, total_sentences):
        """
        计算重复性指标
        """
        if not sentence_stats:
            return {}
        
        metrics = {}
        
        # 1. 整体统计
        total_ethnic_sentences = len(sentence_stats)
        total_ethnic_mentions = sum(s['total_ethnic_count'] for s in sentence_stats)
        
        metrics['total_sentences'] = total_sentences
        metrics['ethnic_sentences_count'] = total_ethnic_sentences
        metrics['ethnic_sentences_ratio'] = total_ethnic_sentences / total_sentences
        metrics['total_ethnic_mentions'] = total_ethnic_mentions
        metrics['mentions_per_sentence'] = total_ethnic_mentions / total_sentences
        
        # 2. 各族群统计
        ethnic_totals = {}
        for ethnic_type in self.ethnic_labels.keys():
            ethnic_total = sum(s['ethnic_counts'].get(ethnic_type, 0) for s in sentence_stats)
            ethnic_totals[ethnic_type] = ethnic_total
        
        metrics['ethnic_distribution'] = ethnic_totals
        
        # 3. 空间分布分析
        ethnic_sentence_indices = [s['sentence_idx'] for s in sentence_stats]
        if len(ethnic_sentence_indices) > 1:
            gaps = np.diff(sorted(ethnic_sentence_indices))
            metrics['average_gap'] = float(np.mean(gaps))
            metrics['gap_std'] = float(np.std(gaps))
            metrics['min_gap'] = int(np.min(gaps))
            metrics['max_gap'] = int(np.max(gaps))
            metrics['clustering_coefficient'] = float(len(gaps[gaps <= 10]) / len(gaps))
        
        # 4. 窗口级别分析
        if window_analysis:
            active_windows = [w for w in window_analysis if w['total_ethnic_mentions'] > 0]
            metrics['active_windows_count'] = len(active_windows)
            metrics['active_windows_ratio'] = len(active_windows) / len(window_analysis)
            
            if active_windows:
                window_diversities = [w['ethnic_diversity'] for w in active_windows]
                metrics['avg_window_diversity'] = float(np.mean(window_diversities))
                metrics['max_window_mentions'] = max(w['total_ethnic_mentions'] for w in active_windows)
        
        # 5. 各族群的重复模式
        for ethnic_type in self.ethnic_labels.keys():
            ethnic_sentences = [s for s in sentence_stats if s['ethnic_counts'].get(ethnic_type, 0) > 0]
            
            if len(ethnic_sentences) > 0:
                indices = [s['sentence_idx'] for s in ethnic_sentences]
                frequency = sum(s['ethnic_counts'].get(ethnic_type, 0) for s in ethnic_sentences)
                
                pattern_info = {
                    'frequency': frequency,
                    'sentence_count': len(ethnic_sentences),
                    'first_appearance': min(indices),
                    'last_appearance': max(indices),
                    'span': max(indices) - min(indices) + 1,
                    'density': len(indices) / (max(indices) - min(indices) + 1) if len(indices) > 1 else 1.0
                }
                
                if len(indices) > 1:
                    gaps = np.diff(sorted(indices))
                    pattern_info.update({
                        'average_gap': float(np.mean(gaps)),
                        'gap_regularity': float(1 / (np.std(gaps) + 1))
                    })
                
                metrics[f'{ethnic_type}_pattern'] = pattern_info
        
        return metrics
    
    def _analyze_temporal_patterns(self, sentence_stats, total_sentences):
        """
        分析时间分布模式
        """
        if not sentence_stats:
            return {}
        
        patterns = {}
        
        # 分成三个时期
        third = total_sentences // 3
        periods = {
            'early': (0, third),
            'middle': (third, 2 * third),
            'late': (2 * third, total_sentences)
        }
        
        period_stats = {}
        for period_name, (start, end) in periods.items():
            period_sentences = [s for s in sentence_stats if start <= s['sentence_idx'] < end]
            
            if period_sentences:
                period_ethnic_counts = {}
                for ethnic_type in self.ethnic_labels.keys():
                    count = sum(s['ethnic_counts'].get(ethnic_type, 0) for s in period_sentences)
                    period_ethnic_counts[ethnic_type] = count
                
                period_stats[period_name] = {
                    'sentence_count': len(period_sentences),
                    'total_mentions': sum(period_ethnic_counts.values()),
                    'ethnic_counts': period_ethnic_counts
                }
            else:
                period_stats[period_name] = {
                    'sentence_count': 0,
                    'total_mentions': 0,
                    'ethnic_counts': {ethnic: 0 for ethnic in self.ethnic_labels.keys()}
                }
        
        patterns['period_distribution'] = period_stats
        
        # 计算时间趋势
        for ethnic_type in self.ethnic_labels.keys():
            ethnic_sentences = [s for s in sentence_stats if s['ethnic_counts'].get(ethnic_type, 0) > 0]
            
            if ethnic_sentences:
                positions = [s['sentence_idx'] / total_sentences for s in ethnic_sentences]
                patterns[f'{ethnic_type}_temporal'] = {
                    'relative_positions': positions,
                    'mean_position': float(np.mean(positions)),
                    'position_spread': float(np.std(positions))
                }
        
        return patterns
    
    def visualize_ethnic_patterns(self, analysis_results):
        """
        可视化族群意象模式
        """
        if not analysis_results:
            print("❌ 无分析结果可视化")
            return
        
        book_title = analysis_results['book_title']
        window_analysis = analysis_results['window_analysis']
        sentence_stats = analysis_results['sentence_level_stats']
        
        if not window_analysis or not sentence_stats:
            print("❌ 无足够数据进行可视化")
            return
        
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle(f'《{book_title}》族群意象重复性分析', fontsize=16, fontweight='bold')
        
        # 1. 滑动窗口中各族群的变化
        ax1 = axes[0, 0]
        window_df = pd.DataFrame(window_analysis)
        
        ethnic_types = list(self.ethnic_labels.keys())
        colors = ["#CF3030", "#CC8F13", "#4149E9", "#067317"]
        
        for i, ethnic_type in enumerate(ethnic_types):
            col_name = f'{ethnic_type}_count'
            if col_name in window_df.columns:
                ax1.plot(window_df['window_center'], window_df[col_name], 
                        marker='o', label=ethnic_type, color=colors[i], linewidth=2, markersize=4)
        
        ax1.set_xlabel('句子位置')
        ax1.set_ylabel('意象出现次数')
        ax1.set_title('滑动窗口族群意象变化')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 2. 族群意象的整体分布
        ax2 = axes[0, 1]
        ethnic_totals = analysis_results['repeatness_metrics'].get('ethnic_distribution', {})
        
        if ethnic_totals and sum(ethnic_totals.values()) > 0:
            # 只显示非零的族群
            non_zero_counts = {k: v for k, v in ethnic_totals.items() if v > 0}
            if non_zero_counts:
                colors_pie = colors[:len(non_zero_counts)]
                wedges, texts, autotexts = ax2.pie(non_zero_counts.values(), 
                                                  labels=non_zero_counts.keys(), 
                                                  autopct='%1.1f%%', 
                                                  colors=colors_pie, 
                                                  startangle=90)
                ax2.set_title('族群意象整体分布')
            else:
                ax2.text(0.5, 0.5, '无族群意象数据', ha='center', va='center', transform=ax2.transAxes)
                ax2.set_title('族群意象整体分布')
        else:
            ax2.text(0.5, 0.5, '无族群意象数据', ha='center', va='center', transform=ax2.transAxes)
            ax2.set_title('族群意象整体分布')
        
        # 3. 句子级别的族群意象散点图
        ax3 = axes[1, 0]
        
        for i, ethnic_type in enumerate(ethnic_types):
            ethnic_sentences = [s for s in sentence_stats if s['ethnic_counts'].get(ethnic_type, 0) > 0]
            if ethnic_sentences:
                indices = [s['sentence_idx'] for s in ethnic_sentences]
                ax3.scatter(indices, [i] * len(indices), 
                           label=f"{ethnic_type} ({len(indices)})", 
                           color=colors[i], s=30, alpha=0.7)
        
        ax3.set_xlabel('句子位置')
        ax3.set_ylabel('族群类型')
        ax3.set_yticks(range(len(ethnic_types)))
        ax3.set_yticklabels(ethnic_types)
        ax3.set_title('族群意象在文本中的分布')
        ax3.grid(True, alpha=0.3)
        ax3.legend()
        
        # 4. 时期分布对比
        ax4 = axes[1, 1]
        temporal = analysis_results['temporal_patterns'].get('period_distribution', {})
        
        if temporal:
            periods = ['early', 'middle', 'late']
            period_names = ['前期', '中期', '后期']
            
            width = 0.2
            x = np.arange(len(period_names))
            
            for i, ethnic_type in enumerate(ethnic_types):
                counts = [temporal.get(period, {}).get('ethnic_counts', {}).get(ethnic_type, 0) 
                         for period in periods]
                if sum(counts) > 0:  # 只显示有数据的族群
                    ax4.bar(x + i * width, counts, width, 
                           label=ethnic_type, color=colors[i], alpha=0.8)
            
            ax4.set_xlabel('时期')
            ax4.set_ylabel('意象出现次数')
            ax4.set_title('各时期族群意象分布')
            ax4.set_xticks(x + width * 1.5)
            ax4.set_xticklabels(period_names)
            ax4.legend()
            ax4.grid(True, alpha=0.3)
        
        plt.tight_layout()
        
        plt.show()
    
    def print_analysis_summary(self, analysis_results):
        """
        打印分析摘要
        """
        if not analysis_results:
            print("❌ 无分析结果")
            return
        
        book_title = analysis_results['book_title']
        metrics = analysis_results['repeatness_metrics']
        ethnic_words = analysis_results['ethnic_words']
        
        print(f"\n" + "="*60)
        print(f"📊 《{book_title}》族群意象重复性分析报告")
        print("="*60)
        
        # 基本统计
        print(f"\n📈 基本统计:")
        print(f"   • 总句子数: {metrics.get('total_sentences', 0):,}")
        print(f"   • 含族群意象的句子: {metrics.get('ethnic_sentences_count', 0)}")
        print(f"   • 族群意象句子比例: {metrics.get('ethnic_sentences_ratio', 0):.2%}")
        print(f"   • 族群意象总出现次数: {metrics.get('total_ethnic_mentions', 0)}")
        print(f"   • 平均每句意象密度: {metrics.get('mentions_per_sentence', 0):.3f}")
        
        # 词汇统计
        print(f"\n📝 各族群词汇库:")
        for ethnic_type, words in ethnic_words.items():
            print(f"   • {ethnic_type}: {len(words)}个词汇")
            if words:
                examples = words[:5]  # 显示前5个例子
                print(f"     词汇: {', '.join(examples)}")
        
        # 族群分布
        print(f"\n👥 族群意象分布:")
        ethnic_dist = metrics.get('ethnic_distribution', {})
        total_mentions = sum(ethnic_dist.values())
        
        for ethnic_type, count in sorted(ethnic_dist.items(), key=lambda x: x[1], reverse=True):
            if count > 0:
                percentage = (count / total_mentions) * 100 if total_mentions > 0 else 0
                print(f"   • {ethnic_type}: {count}次 ({percentage:.1f}%)")
        
        # 重复性模式
        print(f"\n🔄 重复性模式:")
        if 'average_gap' in metrics:
            print(f"   • 平均间隔: {metrics['average_gap']:.1f}句")
            print(f"   • 间隔标准差: {metrics.get('gap_std', 0):.1f}")
            print(f"   • 最小间隔: {metrics.get('min_gap', 0)}句")
            print(f"   • 最大间隔: {metrics.get('max_gap', 0)}句")
            print(f"   • 聚集系数: {metrics.get('clustering_coefficient', 0):.2f}")
        
        # 窗口分析
        if 'active_windows_ratio' in metrics:
            print(f"   • 活跃窗口比例: {metrics['active_windows_ratio']:.2%}")
            print(f"   • 平均窗口多样性: {metrics.get('avg_window_diversity', 0):.1f}")
        
        # 各族群详细模式
        print(f"\n🎯 各族群重复模式:")
        for ethnic_type in self.ethnic_labels.keys():
            pattern_key = f'{ethnic_type}_pattern'
            if pattern_key in metrics:
                pattern = metrics[pattern_key]
                print(f"   • {ethnic_type}:")
                print(f"     - 总频次: {pattern['frequency']}")
                print(f"     - 出现句数: {pattern['sentence_count']}")
                print(f"     - 首次出现: 第{pattern['first_appearance']+1}句")
                print(f"     - 最后出现: 第{pattern['last_appearance']+1}句")
                print(f"     - 分布密度: {pattern['density']:.3f}")
                if 'average_gap' in pattern:
                    print(f"     - 平均间隔: {pattern['average_gap']:.1f}句")
                    print(f"     - 规律性: {pattern['gap_regularity']:.2f}")

# 使用示例函数
def analyze_ethnic_patterns_demo(df_ima, df_books, book_title="古都", window_size=5):
    """
    演示函数：分析指定书籍的族群意象模式
    
    Parameters:
    df_ima (DataFrame): 意象数据
    df_books (DataFrame): 书籍数据  
    book_title (str): 要分析的书名
    window_size (int): 滑动窗口大小
    """
    # 初始化分析器
    analyzer = IntraBookEthnicAnalyzer(df_ima, df_books)
    
    # 进行分析
    results = analyzer.analyze_ethnic_patterns_in_book(book_title, window_size)
    
    if results:
        # 打印分析摘要
        analyzer.print_analysis_summary(results)
        
        # 可视化
        analyzer.visualize_ethnic_patterns(results, save_plot=True)
        
        return results
    else:
        print(f"❌ 无法分析书籍: {book_title}")
        return None

# 主执行代码
if __name__ == "__main__":
    print("族群意象重复性分析器 (基于文本匹配)")
    print("="*50)
    print("统计方法:")
    print("1. 同时以 book 和 label 为筛选条件获取词汇")
    print("2. 在每个窗口中使用文本匹配统计词汇出现次数")
    print("3. 不依赖 idx_in_text 字段")
    print("\n使用方法:")
    print("results = analyze_ethnic_patterns_demo(df_ima, df_books, '古都', window_size=5)")
```
