```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
import jieba
import re

class EmotionDiscontinuityAnalyzer:
    def __init__(self, emotion_bank_df, books_df):
        """
        初始化情感不连续性分析器
        
        Args:
            emotion_bank_df: ChineseEmoBank数据，包含Word, Valence_Mean, Valence_SD列
            books_df: 书籍数据，包含text_chunk_smallest列
        """
        self.emotion_bank = emotion_bank_df
        self.books_df = books_df
        
        # 创建词汇到情感分数的映射
        self.word_to_emotion = {}
        for _, row in self.emotion_bank.iterrows():
            word = row['Word']
            valence = row['Valence_Mean']
            self.word_to_emotion[word] = valence
    
    def calculate_sentence_emotion(self, sentence, method='mean'):
        """
        计算单个句子的情感分数
        
        Args:
            sentence: 输入句子
            method: 计算方法 ('mean', 'weighted_mean', 'sum')
        
        Returns:
            情感分数，如果没有找到情感词则返回5.0（中性）
        """
        # 分词
        words = jieba.lcut(sentence)
        
        emotion_scores = []
        emotion_weights = []  # 基于SD计算权重
        
        for word in words:
            if word in self.word_to_emotion:
                emotion_scores.append(self.word_to_emotion[word])
                
                # 找到对应的SD值作为权重（SD越小，权重越大）
                word_info = self.emotion_bank[self.emotion_bank['Word'] == word]
                if not word_info.empty:
                    sd = word_info.iloc[0]['Valence_SD']
                    # 权重 = 1 / (SD + 0.1)，避免除零
                    weight = 1 / (sd + 0.1)
                    emotion_weights.append(weight)
                else:
                    emotion_weights.append(1.0)
        
        if not emotion_scores:
            return 5.0  # 中性分数
        
        if method == 'mean':
            return np.mean(emotion_scores)
        elif method == 'weighted_mean':
            return np.average(emotion_scores, weights=emotion_weights)
        elif method == 'sum':
            return np.sum(emotion_scores)
        else:
            return np.mean(emotion_scores)
    
    def calculate_text_emotion_sequence(self, text_chunks, window_size=1):
        """
        计算文本块序列的情感分数
        
        Args:
            text_chunks: 文本块列表
            window_size: 窗口大小，1表示逐句分析
        
        Returns:
            情感分数序列
        """
        emotion_sequence = []
        
        for i in range(0, len(text_chunks), window_size):
            # 获取当前窗口的文本
            window_text = ''.join(text_chunks[i:i+window_size])
            emotion_score = self.calculate_sentence_emotion(window_text)
            emotion_sequence.append(emotion_score)
        
        return emotion_sequence
    
    def calculate_discontinuity_metrics(self, emotion_sequence):
        """
        计算不连续性指标
        
        Args:
            emotion_sequence: 情感分数序列
        
        Returns:
            包含各种不连续性指标的字典
        """
        if len(emotion_sequence) < 2:
            return {
                'mean_absolute_change': 0,
                'std_change': 0,
                'max_change': 0,
                'discontinuity_peaks': 0,
                'total_variation': 0
            }
        
        # 计算相邻情感分数的差异
        changes = np.diff(emotion_sequence)
        absolute_changes = np.abs(changes)
        
        # 各种不连续性指标
        metrics = {
            'mean_absolute_change': np.mean(absolute_changes),  # 平均绝对变化
            'std_change': np.std(changes),  # 变化的标准差
            'max_change': np.max(absolute_changes),  # 最大变化
            'total_variation': np.sum(absolute_changes),  # 总变异
            'discontinuity_peaks': 0  # 不连续性峰值数量
        }
        
        # 计算不连续性峰值（变化超过阈值的次数）
        threshold = np.mean(absolute_changes) + 2 * np.std(absolute_changes)
        metrics['discontinuity_peaks'] = np.sum(absolute_changes > threshold)
        
        return metrics
    
    def analyze_book_discontinuity(self, book_index, window_size=1, plot=True):
        """
        分析单本书的情感不连续性
        
        Args:
            book_index: 书籍在DataFrame中的索引
            window_size: 分析窗口大小
            plot: 是否绘制图表
        
        Returns:
            分析结果字典
        """
        book_row = self.books_df.iloc[book_index]
        book_title = book_row['book']
        text_chunks = book_row['text_chunk_smallest']
        
        # 计算情感序列
        emotion_sequence = self.calculate_text_emotion_sequence(text_chunks, window_size)
        
        # 计算不连续性指标
        discontinuity_metrics = self.calculate_discontinuity_metrics(emotion_sequence)
        
        results = {
            'book_title': book_title,
            'emotion_sequence': emotion_sequence,
            'discontinuity_metrics': discontinuity_metrics,
            'total_chunks': len(text_chunks),
            'emotion_chunks': len(emotion_sequence)
        }
        
        if plot:
            self.plot_emotion_sequence(results)
        
        return results
    
    def analyze_all_books(self, window_size=1):
        """
        分析所有书籍的情感不连续性
        
        Args:
            window_size: 分析窗口大小
        
        Returns:
            所有书籍的分析结果列表
        """
        all_results = []
        
        for i in range(len(self.books_df)):
            print(f"正在分析第 {i+1}/{len(self.books_df)} 本书...")
            result = self.analyze_book_discontinuity(i, window_size, plot=False)
            all_results.append(result)
        
        return all_results
    
    def plot_emotion_sequence(self, analysis_result):
        """
        绘制情感序列图
        """
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
        
        emotion_sequence = analysis_result['emotion_sequence']
        book_title = analysis_result['book_title']
        
        # 情感分数序列图
        ax1.plot(emotion_sequence, linewidth=1.5, alpha=0.8)
        ax1.axhline(y=5.0, color='red', linestyle='--', alpha=0.5, label='中性线(5.0)')
        ax1.set_title(f'{book_title} - 情感分数序列')
        ax1.set_xlabel('文本块序号')
        ax1.set_ylabel('情感分数 (1-10)')
        ax1.legend()
        ax1.grid(True, alpha=0.3)
        
        # 情感变化图
        if len(emotion_sequence) > 1:
            changes = np.diff(emotion_sequence)
            ax2.plot(changes, color='orange', linewidth=1.5, alpha=0.8)
            ax2.axhline(y=0, color='red', linestyle='--', alpha=0.5)
            ax2.set_title(f'{book_title} - 情感变化序列')
            ax2.set_xlabel('文本块序号')
            ax2.set_ylabel('情感变化')
            ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
    
    def compare_books_discontinuity(self, analysis_results):
        """
        比较不同书籍的不连续性
        """
        # 提取比较数据
        book_names = [result['book_title'] for result in analysis_results]
        mean_changes = [result['discontinuity_metrics']['mean_absolute_change'] 
                       for result in analysis_results]
        std_changes = [result['discontinuity_metrics']['std_change'] 
                      for result in analysis_results]
        max_changes = [result['discontinuity_metrics']['max_change'] 
                      for result in analysis_results]
        discontinuity_peaks = [result['discontinuity_metrics']['discontinuity_peaks'] 
                              for result in analysis_results]
        
        # 创建比较DataFrame
        comparison_df = pd.DataFrame({
            'book': book_names,
            'mean_absolute_change': mean_changes,
            'std_change': std_changes,
            'max_change': max_changes,
            'discontinuity_peaks': discontinuity_peaks
        })
        
        # 绘制比较图
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        # 平均绝对变化
        axes[0,0].bar(range(len(book_names)), mean_changes)
        axes[0,0].set_title('平均绝对情感变化')
        axes[0,0].set_xticks(range(len(book_names)))
        axes[0,0].set_xticklabels(book_names, rotation=45, ha='right')
        
        # 变化标准差
        axes[0,1].bar(range(len(book_names)), std_changes, color='orange')
        axes[0,1].set_title('情感变化标准差')
        axes[0,1].set_xticks(range(len(book_names)))
        axes[0,1].set_xticklabels(book_names, rotation=45, ha='right')
        
        # 最大变化
        axes[1,0].bar(range(len(book_names)), max_changes, color='red')
        axes[1,0].set_title('最大情感变化')
        axes[1,0].set_xticks(range(len(book_names)))
        axes[1,0].set_xticklabels(book_names, rotation=45, ha='right')
        
        # 不连续性峰值
        axes[1,1].bar(range(len(book_names)), discontinuity_peaks, color='green')
        axes[1,1].set_title('不连续性峰值数量')
        axes[1,1].set_xticks(range(len(book_names)))
        axes[1,1].set_xticklabels(book_names, rotation=45, ha='right')
        
        plt.tight_layout()
        plt.show()
        
        return comparison_df

# 使用示例
"""
# 1. 加载数据
emotion_bank_df = pd.read_csv('chinese_emobank.csv')
books_df = pd.read_csv('books_data.csv')

# 2. 创建分析器
analyzer = EmotionDiscontinuityAnalyzer(emotion_bank_df, books_df)

# 3. 分析单本书
single_result = analyzer.analyze_book_discontinuity(0, window_size=1, plot=True)
print("单本书分析结果:")
print(single_result['discontinuity_metrics'])

# 4. 分析所有书籍
all_results = analyzer.analyze_all_books(window_size=1)

# 5. 比较所有书籍
comparison_df = analyzer.compare_books_discontinuity(all_results)
print("书籍比较结果:")
print(comparison_df)

# 6. 保存结果
comparison_df.to_csv('books_discontinuity_analysis.csv', index=False)
"""
```
